```python
def conv(image, kernel):
    Hi, Wi = image.shape
    Hk, Wk = kernel.shape
    out = np.zeros((Hi, Wi))
    pad_width0 = Hk // 2
    pad_width1 = Wk // 2
    pad_width = ((pad_width0, pad_width0),(pad_width1,pad_width1))
    padded = np.pad(image, pad_width, mode='edge')
    for i in range(Hi):
        for j in range(Wi):
            out[i][j] = np.multiply(kernel, padded[i:i+Hk, j:j+Wk]).sum()
    return out
def gaussian_kernel(size, sigma):
    kernel = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            kernel[i][j] = 1.0/(2*np.pi*np.power(sigma,2))*(np.exp(-1*(np.power(i-size/2,2)+np.power(j-size/2,2))/(2*np.power(sigma,2))))
    return kernel
```

```python
def partial_x(img):
    out = None
    kernel = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    imgHeight, imgWidth = img.shape
    out = np.zeros((imgHeight, imgWidth))
    out = conv(img, kernel)

    return out
def partial_y(img):
    out = None
    kernel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    imgHeight, imgWidth = img.shape
    out = np.zeros((imgHeight, imgWidth))
    out = conv(img, kernel)

    return out
```

```python
def gradient(img):
    G = np.zeros(img.shape)
    theta = np.zeros(img.shape)
    Gx = partial_x(img)
    Gy = partial_y(img)
    print(Gx.shape)
    print(Gy.shape)
    G = np.sqrt(np.power(Gx,2)+np.power(Gy,2))
    theta = np.arctan2(Gy, Gx) * 180 / np.pi
    return G, theta
```

```python
def non_maximum_suppression(G, theta):
    H, W = G.shape
    out = np.zeros((H, W))
    theta = np.floor((theta+22.5)/45)*45
    for i in range(H):
        for j in range(W):
            if theta[i][j] <0:
                theta[i][j] += 360
            if ((j+1)<W) and ((j-1)>=0) and ((i+1)<H) and ((i-1)>=0):
                if (theta[i][j]== 360 or theta[i][j] == 0) or (theta[i][j] == 180):
                    if theta[i][j] >= theta[i][j+1] and theta[i][j] >= theta[i][j-1]:
                        out[i][j] = theta[i][j]
                if (theta[i][j] == 45) or (theta[i][j] == 225):
                    if theta[i][j] >= theta[i-1][j+1] and theta[i][j] >= theta[i+1][j-1]:
                        out[i][j] = theta[i][j]
                if (theta[i][j] == 90) or (theta[i][j] == 270):
                    if theta[i][j] >= theta[i-1][j] and theta[i][j] >= theta[i+1][j]:
                        out[i][j] = theta[i][j]
                if (theta[i][j] == 135) or (theta[i][j] == 315):
                    if theta[i][j] >= theta[i-1][j-1] and theta[i][j] >= theta[i+1][j+1]:
                        out[i][j] = theta[i][j]
    return out
                        
```

```python
def double_thresholding(img, high, low):
    strong_edges = np.zeros(img.shape)
    weak_edges = np.zeros(img.shape)
    strong_edges = img>high
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i][j]<=high and img[i][j]>=low:
                weak_edges[i][j] = 1
    weak_edges = weak_edges>0
    print(weak_edges)
    return strong_edges, weak_edges
```
