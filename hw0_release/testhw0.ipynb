```python
def svd(matrix):
    u = None
    s = None
    v = None
    u, s, v = np.linalg.svd(matrix)
    return u, s ,v
def get_singular_values(matrix, n):
    singular_values = []
    u,s,v = svd(matrix)
    if n>len(s):
        print("wrong number")
        return 0
    for i in range(n):
        singular_values.append(s[i])
    return singular_values
    
```

```python
def eigen_decomp(matrix):
    w = None
    v = None
    w, v = np.linalg.eig(matrix)
    return w,v
def get_eigen_values_and_vectors(matrix, num_values):
    w,v = eigen_decomp(matrix)
    if num_values>len(w):
        print("wrong number")
        return 0
    eigen_values = []
    eigen_vectors = []
    for i in range(num_values):
        eigen_values.append(w[i])
        eigen_vectors.append(v[i])
    return eigen_values, eigen_vectors
```

```python
M = np.matrix([[1,2,3],[4,5,6],[7,8,9]])
val, vec = get_eigen_values_and_vectors(M[:,:3], 1)
print("Values = \n", val)
print("Vectors = \n", vec)
val, vec = get_eigen_values_and_vectors(M[:,:3], 2)
print("Values = \n", val)
print("Vectors = \n", vec)
```

```python
def rgb_decomposition(image, channel):
    if channel == 'R':
        redImg = image[:,:,2]
        return redImg
    elif channel == 'G':
        greenImg = image[:,:,1]
        return greenImg
    else:
        blueImg = image[:,:,0]
        return blueImg
```

```python
without_red = rgb_decomposition(image1, 'R')
without_blue = rgb_decomposition(image1, 'B')
without_green = rgb_decomposition(image1, 'G')
display(image1)
display(without_red)
display(without_blue)
display(without_green)
```

```python
def lab_decomposition(image, channel):
    lab = color.rgb2lab(image)
    if channel == 'L':
        return lab[:,:,2]
    elif channel == 'A':
        return lab[:,:,1]
    elif:
        return lab[:,:,0]
```
